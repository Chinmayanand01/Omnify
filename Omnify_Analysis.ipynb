```python
#Reading data

import pandas as pd
df = pd.read_excel("Omnify-Analyst-Intership-Task.xlsx")
```

```python
# Convert the 'DateColumn' to a pandas datetime object
df['Week'] = pd.to_datetime(df['Week'])

# Extract the month and create a new column 'Month'
df['Month'] = df['Week'].dt.month
```

```python
# Extract the week  and create a new column week

df['Week'] = df['Week'].dt.week
```

```python
df
```

```python
month_names = {
    1: 'January',
    2: 'February',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'October',
    11: 'November',
    12: 'December',
}

# Map the month numbers to month names and create a new column 'MonthName'
df['Month'] = df['Month'].map(month_names)
df
```

```python
df = df.rename(columns={'Cost ($)': 'Spends', 'Payment ($)': 'Revenue'})
df
```

```python
df.columns
```

```python
import pandas as pd

# Convert "Payment Date" to datetime if it's not already in the proper format
df["Payment Date"] = pd.to_datetime(df["Payment Date"])

# Convert other numeric columns to appropriate data types (if needed)
df["Clicks"] = df["Clicks"].astype(int)
df["Spends"] = df["Spends"].astype(float)
df["Leads"] = df["Leads"].astype(int)
df["Revenue"] = df["Revenue"].str.replace('$', '')
df["Revenue"] = df["Revenue"].str.replace(',', '').astype(float)
df["Revenue"] = df["Revenue"].astype(float)

# Set "Payment Date" as the index of the DataFrame
df.set_index("Payment Date", inplace=True)

# Now, you can perform various time series analysis tasks as shown in the previous code snippet.
```

```python

import pandas as pd

monthly_data = df.resample('M').sum(numeric_only=True)

monthly_stats = df.resample('M').agg({'Clicks': 'mean', 'Spends': 'sum', 'Leads': 'sum'})

monthly_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Plotting a scatter plot between "Clicks" and "Spends" columns
plt.figure(figsize=(10, 6))
plt.scatter(df['Country'], df['Revenue'])
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.title('Scatter Plot: Clicks vs Spends')
plt.show()
```

```python
# Group the data by "Ad group" and sum the "Clicks" for each ad group
adgroup_clicks = df.groupby("Ad group")["Clicks"].sum()

# Plotting a bar graph between "Ad group" and "Clicks" columns
plt.figure(figsize=(10, 6))
adgroup_clicks.plot(kind="bar")
plt.xlabel('Ad group')
plt.ylabel('Clicks')
plt.title('Bar Graph: Clicks by Ad group')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Group the data by "Campaign" and sum the "Spends" for each campaign
campaign_spends = df.groupby("Campaign")["Spends"].sum()

# Plotting a bar graph between "Campaign" and "Spends" columns
plt.figure(figsize=(10, 6))
campaign_spends.plot(kind="bar")
plt.xlabel('Campaign')
plt.ylabel('Spends')
plt.title('Bar Graph: Spends by Campaign')
plt.show()
```

```python
#Time series analysis

import pandas as pd
import matplotlib.pyplot as plt


# Resample and sum the numeric columns for each month
monthly_data = df.resample('M').sum(numeric_only=True)

# Plotting the time series of "Clicks" and "Spends" over months
plt.figure(figsize=(10, 6))
plt.plot(monthly_data.index, monthly_data['Clicks'], label='Clicks')
plt.plot(monthly_data.index, monthly_data['Spends'], label='Spends')
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Time Series: Clicks and Spends over Months')
plt.legend()
plt.show()
```

```python
# Calculate rolling mean over a window of 3 months
rolling_mean = monthly_data['Spends'].rolling(window=3).mean()

# Plot the original time series and the rolling mean
plt.figure(figsize=(10, 6))
plt.plot(monthly_data['Spends'], label='Original')
plt.plot(rolling_mean, label='Rolling Mean (3 months)')
plt.xlabel('Date')
plt.ylabel('Spends')
plt.title('Time Series with Rolling Mean')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA


# Resample and sum the numeric columns for each month
monthly_data = df.resample('M').sum(numeric_only=True)

# Fit ARIMA model and make predictions
model = ARIMA(monthly_data['Spends'], order=(1, 1, 1))
model_fit = model.fit()
forecast = model_fit.forecast(steps=12)  # Forecasting 12 months into the future

# Plot the original time series and the forecasted values
plt.figure(figsize=(10, 6))
plt.plot(monthly_data['Spends'], label='Original')
plt.plot(forecast, label='Forecast')
plt.xlabel('Date')
plt.ylabel('Spends')
plt.title('Time Series Forecasting with ARIMA')
plt.legend()
plt.show()
```
